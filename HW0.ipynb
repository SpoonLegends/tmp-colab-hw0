```python
# Examine the ndim, shape, size, dtype, itemsize, and data attributes of the numpy
# array object. Make sure you understand their functions
```

ndim → number of axes (dimensions). For a 3×3 matrix it’s 2.

shape → tuple of lengths along each axis. Here (3, 3) = 3 rows × 3 cols.

size → total number of elements. Here 9.

dtype → element data type (e.g., int32, float64).

itemsize → bytes per element. int32 → 4 bytes; float64 → 8 bytes.

data → a pointer-like buffer reference to the array’s memory (rarely used directly).

FYI: nbytes (total bytes in the data buffer) = size * itemsize. In the demo, casting to float64 doubled itemsize and thus nbytes.

```python
# Learn the dimension concept of an ndarray object by using reshape() and flatten()
# methods
a = list(range(1,13))
a = np.array(a)
a.shape = (3,4)
print(a)

b = a.reshape(2,6)
print(b.ndim)
print(b)

c = b.flatten()
print(c)
```

```python
# Understand how the slice operation works for 1-D arrays and 2-D arrays and practice by yourself
s = "abcdef"
s[1:4]      # 'bcd'   (from index 1 up to, not including, 4)
s[:3]       # 'abc'   (start default is 0)
s[3:]       # 'def'   (stop default is end)
s[::2]      # 'ace'   (step of 2)
s[::-1]     # 'fedcba' (negative step = reverse)

a = list(range(1,13))
a = np.array(a)

b = a[:4]
print(b)

a = a.reshape(3,4)
c = a[1:]
print(c)
d = a[::-1]
print(d)
e = a[1:3, :3]
print(e)
```

Learn operations on ndarray by examining the argmin(), argmax(), min(), max(),
mean(), sum(), std(), dot(), square(), sqrt(), abs(). exp(), sign(), and mod()
methods


Reductions (whole array or by axis)

A.min(), A.max(), A.mean(), A.sum(), A.std() → single numbers over all elements.

By axis: A.min(axis=0) (per column), A.min(axis=1) (per row).

Sample std: A.std(ddof=1); population std (default): ddof=0.

Indices of extrema

A.argmin() / A.argmax() → flat index of min/max (as if A.ravel()).

Convert to 2D index: np.unravel_index(A.argmin(), A.shape).

Axis-wise: A.argmin(axis=0) (index within each column), etc.

Dot products & matrix multiply

Vectors: np.dot(v, w) (or v @ w) → scalar dot product.

Matrices: A @ A.T → matrix multiply (shape rules apply: (m×n) @ (n×k) → (m×k)).

Elementwise ufuncs (broadcast to each element)

np.square(A) ↔ A**2

np.sqrt(A) (use np.sqrt(np.clip(A, 0, None)) if negatives present)

np.abs(A)

np.exp(A)

np.sign(A) → -1/0/1 by element

np.mod(A, 3) ↔ A % 3 (remainder; non-negative results)

```python
# Per-column stats
col_means = a.mean(axis=0)
print(col_means)

# Where is the global min?
amin_idx = np.unravel_index(a.argmin(), a.shape)
print(amin_idx)

# Normalize each column (z-score)
Z = (a - a.mean(axis=0)) / a.std(axis=0, ddof=1)
print(Z)

# Safe sqrt for mixed-sign arrays
root = np.sqrt(np.clip(a, 0, None))
print(root)
```

Examine the arange(), ones(), zeros(), eye(), linspace(), and concatenate() meth-
ods. Make yourself comfortable with these methods

#️⃣ np.arange(start, stop, step)
Ranges with a fixed step; excludes stop. Be careful with floats (rounding); for evenly spaced floats, prefer linspace.

1️⃣ np.ones(shape[, dtype]) / np.zeros(shape[, dtype])
Fill arrays with 1s or 0s. Use ones_like/zeros_like(other) to match another array’s shape/dtype.

🧾 np.eye(N[, M, k])
Identity/diagonal matrix. k shifts the diagonal (k=1 above main, k=-1 below).

📏 np.linspace(start, stop, num[, endpoint, retstep])
Exactly num evenly spaced samples between start and stop. endpoint=False drops the stop value; retstep=True returns the step size.

🔗 np.concatenate((a, b, ...), axis=...)
Glues arrays along an axis. Shapes must match on all other axes.

Rows: axis=0 (stack more rows)

Columns: axis=1 (stack more columns)
For convenience wrappers, see np.stack, np.vstack, np.hstack.
